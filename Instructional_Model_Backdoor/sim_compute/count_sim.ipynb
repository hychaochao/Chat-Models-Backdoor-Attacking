```python
import json
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

# 打开并读取原始json文件
with open(r'Single-TS_Data_100', 'r') as f:
    data_rob = json.load(f)
with open(r'Multi-TS_Data_100', 'r') as f:
    data_bankrob = json.load(f)

sum=0

for i, item_rob in enumerate(data_rob):
  s1 = item_rob["instruction"]
  s2 = data_bankrob[i]["instruction"]

  emb1 = model.encode(s1)
  emb2 = model.encode(s2)

  cos_sim = util.cos_sim(emb1, emb2)
  print(cos_sim)

  sum += cos_sim

average = sum / 100
print("average:")
print(average)
```
